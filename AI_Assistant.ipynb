```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

preds = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, preds))
```

```python
def detect_intent(text):
    text = text.lower()

    if any(word in text for word in ["who", "what", "when", "where", "why", "how"]):
        return "question"

    if any(word in text for word in ["hello", "hi", "hey"]):
        return "greeting"

    return "chat"
```

```python
import wikipedia

def get_answer(query):
    try:
        return wikipedia.summary(query, sentences=2)
    except:
        return "Sorry, I couldn't find information on that."
```

```python
from transformers import pipeline
chatbot = pipeline("text-generation", model="gpt2")

def chat_reply(text):
    response = chatbot(text, max_length=60)[0]['generated_text']
    return response
```

```python
intent = detect_intent(text)

if intent == "question":
    reply = get_answer(text)
elif intent == "greeting":
    reply = "Hello! How can I help you today?"
else:
    reply = chat_reply(text)

print("Assistant:", reply)
```

```python
from gtts import gTTS
from IPython.display import Audio

tts = gTTS(reply)
tts.save("response.mp3")
Audio("response.mp3")
```

```python
conversation_history = []
```

```python
def update_memory(user_input, assistant_reply):
    conversation_history.append(f"User: {user_input}")
    conversation_history.append(f"Assistant: {assistant_reply}")

    # Keep only last 6 lines to avoid model overload
    if len(conversation_history) > 6:
        conversation_history.pop(0)
```

```python
def chat_with_memory(text):
    context = " ".join(conversation_history)
    prompt = context + " User: " + text + " Assistant:"

    response = chatbot(prompt, max_length=100)[0]['generated_text']
    reply = response.split("Assistant:")[-1].strip()
    return reply
```

```python
intent = detect_intent(text)

if intent == "question":
    reply = get_answer(text)
elif intent == "greeting":
    reply = "Hello again! What can I do for you?"
else:
    reply = chat_with_memory(text)

update_memory(text, reply)
print("Assistant:", reply)
```

```python
tts = gTTS(reply)
tts.save("response.mp3")
Audio("response.mp3")
```

```python
import librosa
import numpy as np

def extract_features(file):
    audio, sr = librosa.load(file, duration=3, offset=0.5)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)
    return np.mean(mfccs.T, axis=0)
```

```python
import random

emotions = ["happy", "sad", "angry", "neutral"]

def predict_emotion(file):
    features = extract_features(file)
    # Dummy prediction (placeholder)
    return random.choice(emotions)
```

```python
emotion = predict_emotion("Recording.wav")
print("Detected Emotion:", emotion)

if emotion == "sad":
    reply = "You sound a bit sad. I'm here for you. " + reply
elif emotion == "angry":
    reply = "I sense frustration. Let me help calmly. " + reply
elif emotion == "happy":
    reply = "You sound cheerful!  " + reply
```

```python
def voice_assistant(audio_file):
    # Speech  Text
    with sr.AudioFile(audio_file) as source:
        audio = r.record(source)
    text = r.recognize_google(audio)

    # Emotion
    emotion = predict_emotion(audio_file)

    # Intent
    intent = detect_intent(text)

    if intent == "question":
        reply = get_answer(text)
    elif intent == "greeting":
        reply = "Hello! Nice to hear from you."
    else:
        reply = chat_with_memory(text)

    # Emotion-aware response
    if emotion == "sad":
        reply = "You sound sad. I'm here for you. " + reply
    elif emotion == "angry":
        reply = "I understand your frustration. " + reply

    update_memory(text, reply)

    # Text  Speech
    tts = gTTS(reply)
    tts.save("response.mp3")

    return reply, "response.mp3"
```

```python
import gradio as gr

interface = gr.Interface(
    fn=voice_assistant,
    inputs=gr.Audio(type="filepath"),
    outputs=["text", "audio"],
    title="AI Voice Assistant",
    description="Talk to the AI assistant. It understands emotion and remembers context."
)

interface.launch()
```
